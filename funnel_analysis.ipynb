```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")

df.head()
```

```python
funnel_data = {
    "Stage": ["Total Customers", "Active Customers", "Churned Customers"],
    "Users": [
        len(df),
        len(df[df['Churn'] == 'No']),
        len(df[df['Churn'] == 'Yes'])
    ]
}

funnel_df = pd.DataFrame(funnel_data)
funnel_df
```

```python
plt.figure(figsize=(6,4))

plt.bar(funnel_df["Stage"], funnel_df["Users"])

plt.title("Customer Funnel Analysis")
plt.xlabel("Stages")
plt.ylabel("Number of Customers")

plt.show()
```

```python
total = len(df)
active = len(df[df['Churn'] == 'No'])
churned = len(df[df['Churn'] == 'Yes'])

retention_rate = (active / total) * 100
churn_rate = (churned / total) * 100

print("Retention Rate:", retention_rate, "%")
print("Churn Rate:", churn_rate, "%")
```

## Insights

- Majority of customers are retained by the company
- A significant percentage of customers have churned
- Reducing churn can improve overall business performance
- Retention strategies should be prioritized
